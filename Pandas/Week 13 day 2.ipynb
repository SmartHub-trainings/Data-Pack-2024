```python
for continent in drinks.continent.unique():
    contient_drinks =drinks[drinks.continent==continent]
    continent_consumption=contient_drinks.total_litres_of_pure_alcohol.sum()
    print(f"{continent}: {continent_consumption}")
```

```python
consumptions=[]
for continent in drinks.continent.unique():
    contient_drinks =drinks[drinks.continent==continent]
    continent_consumption=contient_drinks.total_litres_of_pure_alcohol.sum()
    print(f"{continent}: {continent_consumption}")
    consumptions.append(continent_consumption)
print(sorted(consumptions))
```
