```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
```

```python
dataset=load_boston()
X,y=load_boston(return_X_y=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(4, 4, figsize=(15, 10))
axes = axes.flatten()  # Flatten in case it's a 2D array

for i in range(X.shape[1]):
    X_use = X[:, i]
    print(dataset.feature_names[i])
    print(i)
    print(np.corrcoef(X_use,y))
    print(str("="*40)+"\n")
    
    axes[i].scatter(X_use, y)
    axes[i].set_title(f"Feature {i+1}")
```

```python
sl_model =LinearRegression()
```

```python
X_13=X[:,12].reshape(-1,1)
```

```python
sl_model.fit(X_13,y)
```

```python
y_pred_13=sl_model.predict(X_13)
```

```python
sl_model.score(X_13,y)
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y,y_pred_13)**0.5
```

```python
rmse
```

```python
plt.scatter(X_13,y)
plt.scatter(X_13,y_pred_13)
```
