```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X =np.linspace(20,100)
y = 4+10*X+50*np.random.randn(50)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
sl_model=LinearRegression()
X=X.reshape(-1,1)
sl_model.fit(X,y)
y_pred_1=sl_model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred_1,"r")
print("Score",r2_score(y,y_pred_1))
print("rmse",mean_squared_error(y,y_pred_1)**0.5)
```

```python
sl_model.coef_
```

```python
sl_model.intercept_
```
