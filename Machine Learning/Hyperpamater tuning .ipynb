```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
X,y= load_breast_cancer(return_X_y=True)
```

```python
def knn_model(X,y,n=3,p=1,weight="uniform"):
    X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10)
    model = KNeighborsClassifier(n_neighbors=n,p=p,weights=weight)
    model.fit(X_train,y_train)
    score =model.score(X_test,y_test)
    params={"weight":weight,"p":p,"n":n,"score":score}
    return params
```

```python
import pandas as pd
```

```python
result_dic={
   "scores":[] ,
    "weights":[],
    "p":[],
    "n":[]
}
for weight in ["uniform","distance"]:
        for p in range(1,5):
            for n in range(3,15,2):
                output=knn_model(X,y,n,p,weight)
                result_dic["scores"].append(output["score"])
                result_dic["p"].append(output["p"])
                result_dic["weights"].append(output["weight"])
                result_dic["n"].append(output["n"])
df=pd.DataFrame(result_dic)
df
```
